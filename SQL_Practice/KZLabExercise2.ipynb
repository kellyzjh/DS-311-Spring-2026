```python
customers_query = \
"""
SELECT City, COUNT(*) AS "number of customers"
FROM customers
GROUP BY City
ORDER BY "number of customers" DESC

"""

with sqlite3.connect('data/chinook.db') as chin_con: 
   customers_df = pd.read_sql(customers_query, chin_con)

customers_df
```

5f) The cities with the largest number of customers are São Paulo, Prague, Paris, Mountain View, London, and Berlin, with 2 customers in each of those countries.	 

```python
genre_query = \
"""
SELECT tracks.GenreId, genres.Name, COUNT(*) AS "number of tracks"
FROM tracks
JOIN genres ON tracks.GenreId = genres.GenreID
GROUP BY tracks.GenreId
ORDER BY "number of tracks" DESC

"""

with sqlite3.connect('data/chinook.db') as chin_con: 
   genre_df = pd.read_sql(genre_query, chin_con)

genre_df
```

```python
import sqlite3
import pandas as pd
```

```python
airports_query = \
"""
SELECT *
FROM airports
"""

with sqlite3.connect('data/flights.db') as con:   
   airports_df = pd.read_sql(airports_query, con)

airports_df
```

```python
airports_query = \
"""
SELECT DISTINCT country AS "airport country"
FROM airports
"""

with sqlite3.connect('data/flights.db') as con:
    airports_df = pd.read_sql(airports_query, con)

airports_df
```

```python
airports_query = \
"""
SELECT COUNT(*) AS "Number of Airports"
FROM airports
"""

with sqlite3.connect('data/flights.db') as con:
    airports_df = pd.read_sql(airports_query, con)

airports_df
```

```python
airports_query = \
"""
SELECT UPPER(name) AS "airport names in caps"
FROM airports
"""

with sqlite3.connect('data/flights.db') as con:
    airports_df = pd.read_sql(airports_query, con)

airports_df
```

```python
airports_query = \
"""
SELECT *
FROM airports
WHERE city = 'New York' OR city = 'Goroka'
"""

with sqlite3.connect('data/flights.db') as con:
    airports_df = pd.read_sql(airports_query, con)

airports_df
```

```python
airports_query = \
"""
SELECT MAX(CAST(altitude as int)) AS "highest altitude"
FROM airports
"""

with sqlite3.connect('data/flights.db') as con:
    airports_df = pd.read_sql(airports_query, con)

airports_df
```

```python
airports_query = \
"""
SELECT name as "airport name",
    CAST(latitude AS int) AS "Airport Latitude",
    CAST(altitude AS int) AS "Altitude(ft)",
    country AS "airport country"
    
FROM airports
WHERE "altitude(ft)" >= 10000 AND "Airport Latitude" Between 20 AND 50 

"""

with sqlite3.connect('data/flights.db') as con:
    airports_df = pd.read_sql(airports_query, con)

airports_df
```
